Dictionary

```python
## Dictionary

# list => array, mengakses dengan menggunakan index

data_list = ["Ucup", "Salsa", "Xinthink"]
# print(data_list,"\n")
# print(data_list[2])

# Dictionary (Dict) => array asosiatif
# identifier => key

data_dict = {
    'cp' : 'Ucup',
    'sl' : 'Salsa',
    'xin' : 'Xinthink',
    'nmb' : 100,
    'list' : data_list,

}
print(data_dict) # untuk memanggil semua data
print(data_dict['sl']) # untuk memanggil salah satu data
print(data_dict['cp']) # untuk memanggil salah satu data
```

Operasi Dictionary

```python
## operasi dictionary

data_dict = {
    "cup" : "ucup surucup",
    "pet" : "opet saripet",
    "ndo" : "nando surundo"
}

# panjang dictionary
dict_len = len(data_dict)
print(f"panjang dictionary : {dict_len}")

# mengcek key exist atau tidak
key = "cup"
check_key = key in data_dict
print(f"apakah {key} ada di data_dict : {check_key}\n")

# mengakses value (read) menggunakan get
print(data_dict["cup"],"\n")
print(data_dict.get("cup"),"\n")
print(data_dict.get("cu","key tersebut tidak ditemukan\n"))

# mengubah data
print(f"sebelum diubah{data_dict}\n")
data_dict["cup"]= "ucup si ganteng"

data_dict["dee"]= "ade si lucu"
print(f"setelah diubah diubah{data_dict}\n")

# update
data_dict.update({"pet":"opet si buhil"})
print(f"sebelum ditambah{data_dict}\n")

data_dict.update({"ren":"ren si oren"})
print(f"Setelah ditambah{data_dict}\n")

# menghapus data
del data_dict["ren"]
print(f"setelah dihapus :{data_dict}")



```

Looping Dictionary

```python
## Looping
teman_teman = {
    "cup" : "ucup surucup",
    "tong" : "otong surotong",
    "lur" : "lur sedulur",
    "well" : "wel bung towel"
}

# print(teman_teman)

# looping
# for teman in teman_teman:
#   print(teman)

# operator untuk mengambil item
# keys = teman_teman.keys()
# print(keys,"\n")

for key in teman_teman.keys():
  print(key)

values = teman_teman.values()
print(values)

for value in teman_teman.values():
  print(value)

items = teman_teman.items()
print(items)

for item in teman_teman.items():
  print(item)

for key,value in teman_teman.items():
  print(f"key = {key}, valuenya adalah {value}")
```

2. Dictionary

```python
keranjang = {
    "apel" : 5000,
    "pisang" : 3000,
    "jeruk" : 7000,
    "anggur" : 15000
}


total = 0

for buah, harga in keranjang.items():
  print(f"Harga buah {buah} : Rp{harga}")
  total = total + harga
print(f"\nharga total : {total}")
```
